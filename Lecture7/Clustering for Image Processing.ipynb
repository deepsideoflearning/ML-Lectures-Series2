```python
from PIL import Image
from sklearn.cluster import KMeans
```

```python
image = Image.open('dogs.jpeg')
```

```python
from matplotlib import pyplot as plot

%matplotlib inline

plot.imshow(image)
```

```python
import numpy as np

image_np = np.asarray(image)
```

```python
image_np.shape
```

```python
image_np = image_np.reshape((185*272,3))
```

```python
image_np.shape
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(image_np)
```

```python
labels = kmeans.predict(image_np)
```

```python
labels = labels.reshape((185,272))
```

```python
plot.imshow(labels)
```

```python
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
```

```python
blob = np.ones((100,100,3))
```

```python
blob1 = blob * centers[0]
```

```python
plot.imshow(blob1)
```

```python
blob2 = blob * centers[1]
plot.imshow(blob2)
```

```python
blob2=blob2.astype(int)
blob2
```
